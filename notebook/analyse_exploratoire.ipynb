```python
COLS = {
    "event_id": "Identifiant",
    "url": "URL canonique",
    "title": "Titre",
    "desc": "Description",
    "desc_long": "Description longue",
    "keywords": "Mots clés",
    "updated_at": "Dernière mise à jour",
    "first_start": "Première date - Début",
    "first_end": "Première date - Fin",
    "last_start": "Dernière date - Début",
    "last_end": "Dernière date - Fin",
    "state": "État de l'événement",
    "mode": "Événement physique ou en ligne",
    "online_link": "Lien d'accès en ligne",
    "geo": "Coordonnées géographiques",
    "venue_name": "Nom du lieu",
    "address": "Adresse",
    "city": "Ville",
    "postal_code": "Code postal",
    "department": "Département",
    "region": "Région",
    "country": "Pays",
    "country_name": "Nom du pays",
}
```

```python
df2 = df.copy()

date_cols = [COLS["first_start"], COLS["first_end"], COLS["last_start"], COLS["last_end"]]

for c in date_cols:
    # Force un parsing en UTC pour homogénéiser les timezones
    df2[c] = pd.to_datetime(df2[c], errors="coerce", utc=True)

# Choix des bornes globales de l'événement
df2["event_start"] = df2[COLS["first_start"]]
df2["event_end"] = df2[COLS["last_end"]]

# (Optionnel) si tu veux des datetimes "naïfs" (sans timezone)
# df2["event_start"] = df2["event_start"].dt.tz_convert("Europe/Paris").dt.tz_localize(None)
# df2["event_end"] = df2["event_end"].dt.tz_convert("Europe/Paris").dt.tz_localize(None)

print("Missing rates:")
print(df2[date_cols + ["event_start", "event_end"]].isna().mean().round(4))

display(df2[[COLS["event_id"], "event_start", "event_end"]].head(5))
```

```python
def parse_latlon(s: str | float | None):
    if s is None or (isinstance(s, float) and np.isnan(s)):
        return (np.nan, np.nan)
    if not isinstance(s, str):
        s = str(s)
    # extraire deux nombres (avec . ou ,)
    nums = re.findall(r"-?\d+(?:[.,]\d+)?", s)
    if len(nums) < 2:
        return (np.nan, np.nan)
    a = float(nums[0].replace(",", "."))
    b = float(nums[1].replace(",", "."))
    # hypothèse la plus fréquente en France: lat ~ [41..51], lon ~ [-5..10]
    if 41 <= a <= 51 and -10 <= b <= 15:
        return (a, b)  # lat, lon
    if 41 <= b <= 51 and -10 <= a <= 15:
        return (b, a)  # lon,lat -> lat,lon
    return (a, b)

latlon = df2[COLS["geo"]].map(parse_latlon)
df2["lat"] = [x[0] for x in latlon]
df2["lon"] = [x[1] for x in latlon]

print("lat missing:", df2["lat"].isna().mean().round(4))
print("lon missing:", df2["lon"].isna().mean().round(4))
display(df2[[COLS["geo"], "lat", "lon"]].dropna().head(10))
```

```python
import pandas as pd

# bornes tz-aware en UTC (cohérentes avec event_start/event_end)
today = pd.Timestamp.now(tz="UTC").normalize()
past = today - pd.Timedelta(days=365)
future = today + pd.Timedelta(days=365)

print("today:", today, "past:", past, "future:", future)

base = df2.dropna(subset=["event_start"]).copy()

# filtre 1 : ta proposition (probablement très restrictif)
f1 = base[(base["event_start"] >= past) & (base["event_end"] >= future)]

# filtre 2 : fenêtre sur start
f2 = base[(base["event_start"] >= past) & (base["event_start"] <= future)]

# filtre 3 : f2 + événements encore "actifs" (end>=today si end existe)
f3 = f2[(f2["event_end"].isna()) | (f2["event_end"] >= today)]

print("Counts:")
print("f1 (start>=past & end>=future):", len(f1))
print("f2 (past<=start<=future):     ", len(f2))
print("f3 (f2 + end>=today):         ", len(f3))

print("\nState value counts (top 10) on f2:")
display(f2[COLS["state"]].value_counts().head(10))
```

```python
# bornes déjà définies en UTC dans ta cellule D :
# today, past, future (tz-aware UTC)

base = df2.dropna(subset=["event_start"]).copy()
f2 = base[(base["event_start"] >= past) & (base["event_start"] <= future)].copy()

# is_upcoming : event_end >= today (ou True si pas de end)
f2["is_upcoming"] = f2["event_end"].isna() | (f2["event_end"] >= today)

def parse_state(s):
    """Retourne (state_id, state_label_fr) depuis la colonne 'État de l'événement'."""
    if s is None or (isinstance(s, float) and np.isnan(s)):
        return (None, None)
    if not isinstance(s, str):
        s = str(s)
    try:
        obj = json.loads(s)
        sid = obj.get("id")
        label = obj.get("label", {}).get("fr")
        return (sid, label)
    except Exception:
        # fallback très simple
        if '"id": 1' in s:
            return (1, "Programmé")
        if '"id": 6' in s:
            return (6, "Annulé")
        return (None, None)

state_parsed = f2[COLS["state"]].map(parse_state)
f2["state_id"] = [x[0] for x in state_parsed]
f2["state_label_fr"] = [x[1] for x in state_parsed]

print("Rows f2:", len(f2))
print("Upcoming rows:", int(f2["is_upcoming"].sum()))
display(f2[["event_start","event_end","is_upcoming","state_id","state_label_fr"]].head(10))
```

```python
keep = {
    "event_id": COLS["event_id"],
    "url": COLS["url"],
    "title": COLS["title"],
    "description": COLS["desc"],
    "description_long": COLS["desc_long"],
    "keywords": COLS["keywords"],
    "updated_at_raw": COLS["updated_at"],
    "venue_name": COLS["venue_name"],
    "address": COLS["address"],
    "city": COLS["city"],
    "postal_code": COLS["postal_code"],
    "department": COLS["department"],
    "region": COLS["region"],
    "country": COLS["country"],
    "country_name": COLS["country_name"],
    "mode": COLS["mode"],
    "online_link": COLS["online_link"],
}

proc = f2[list(keep.values())].rename(columns={v: k for k, v in keep.items()}).copy()

# ajouter dates + geo + état
proc["event_start"] = f2["event_start"]
proc["event_end"] = f2["event_end"]
proc["lat"] = f2["lat"]
proc["lon"] = f2["lon"]
proc["is_upcoming"] = f2["is_upcoming"]
proc["state_id"] = f2["state_id"]
proc["state_label_fr"] = f2["state_label_fr"]

print(proc.shape)
display(proc.head(5))
```

```python
def clean_text(x):
    if x is None or (isinstance(x, float) and np.isnan(x)):
        return ""
    x = str(x)
    return " ".join(x.split())

def build_doc_text(row):
    title = clean_text(row["title"])
    desc = clean_text(row["description"])
    desc_long = clean_text(row["description_long"])
    kw = clean_text(row["keywords"])
    venue = clean_text(row["venue_name"])
    city = clean_text(row["city"])
    dept = clean_text(row["department"])
    region = clean_text(row["region"])
    url = clean_text(row["url"])

    # on privilégie Description, puis longue si Description vide
    main_desc = desc if desc else desc_long

    # dates (affichage simple)
    start = row["event_start"]
    end = row["event_end"]
    start_s = start.date().isoformat() if pd.notna(start) else ""
    end_s = end.date().isoformat() if pd.notna(end) else ""

    parts = [
        f"Titre: {title}",
        f"Description: {main_desc}" if main_desc else "",
        f"Lieu: {venue}" if venue else "",
        f"Ville: {city}" if city else "",
        f"Département: {dept}" if dept else "",
        f"Région: {region}" if region else "",
        f"Dates: du {start_s} au {end_s}" if start_s else "",
        f"Mots-clés: {kw}" if kw else "",
        f"URL: {url}" if url else "",
    ]
    txt = "\n".join([p for p in parts if p])
    return txt.strip()

proc["doc_text"] = proc.apply(build_doc_text, axis=1)

# Contrôle rapide
print("doc_text empty rate:", (proc["doc_text"].str.len() == 0).mean().round(4))
display(proc[["event_id","title","is_upcoming","doc_text"]].head(3))
```

```python
def norm(s: str) -> str:
    if s is None:
        return ""
    s = str(s).strip().lower()
    s = "".join(ch for ch in unicodedata.normalize("NFKD", s) if not unicodedata.combining(ch))
    s = " ".join(s.split())
    return s

proc2 = proc.copy()
proc2["region_norm"] = proc2["region"].map(norm)

TARGET_REGIONS = {norm("Nouvelle-Aquitaine")}
geo = proc2[proc2["region_norm"].isin(TARGET_REGIONS)].copy()

print("Rows after region filter:", len(geo))
geo["country_name"].value_counts().head(10)
```
